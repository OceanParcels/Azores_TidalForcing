# FTLE plot:

* Day 01 no tides:

```python
fig = plt.figure(figsize=(16, 19)) #(17,5)
gs = gridspec.GridSpec(5, 3, height_ratios=[.24, .24, .24, .24, .01])
axC = plt.subplot(gs[-1, :])

ntitles = nmonths

step = .004 # degrees
grid_lons, grid_lats = np.meshgrid(np.arange(-27, -21+step, step), np.arange(32.5, 36.5+step, step))
    
for ii in range(0, len(ntitles)):
    FTLE_b = dFTLE_nTs_w01[ii]['FTLE_b']
    ax1 = plt.subplot(gs[ii], projection=ccrs.PlateCarree())    
    ax1.set_xlim([subregion_v02.lonmin, subregion_v02.lonmax])
    ax1.set_ylim([subregion_v02.latmin, subregion_v02.latmax])
    cc = ax1.scatter(grid_lons, grid_lats, c=FTLE_b, s=5, vmin=0., vmax=0.7, cmap='viridis', transform=ccrs.PlateCarree()) #marker='o'
    #     ax1.coastlines(resolution='50m')
    #     ax1.add_feature(cartopy.feature.LAND, facecolor='black', zorder=20)
    gl = ax1.gridlines(crs=ccrs.PlateCarree(), draw_labels=True)
    gl.top_labels = False
    gl.right_labels = False
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    gl.ylocator = mticker.FixedLocator([33., 34., 35., 36.])
    #gl.xlocator = mticker.FixedLocator([0, 5, 10, 15, 20, 25, 30, 35])
    #     ax1.set_xticks([-26, -24., -22.])
    #     ax1.set_yticks([])
    
    ######
    # Title
    if ntitles[ii] in ntitles[0:6]:
        nyear = '2009'
    elif ntitles[ii] in ntitles[6::]:
        nyear = '2010'
    else:
        print('Error nyear')
        dgdsgsg
        
    strtitle = ntitles[ii] + ' ' + nyear
    ax1.set_title(strtitle, size=16)
    ######
    
#plt.suptitle(nmonth + ' back - forward', size=20) #+ nyear 
cbar = plt.colorbar(cc, cax=axC, extend='max', orientation='horizontal')
cbar.ax.get_yaxis().labelpad = 15

 plt.savefig('../../Figures/' + 'FTLE_backward_v02_biweekly_AllMonths_day01_nT.jpg', dpi=300)

plt.show()

```
